```python
import numpy as np
import matplotlib.pyplot as plt
```

# Using the following schemes: (i) forward Euler; (ii) backward Euler; (iii) Trapezoidal method; (iv) Second-order Runge-Kutta method.

# Try values of ∆t = 0.2,0.9,1.1,2.5.

```python
def FE(ts, y0):
    dt = ts[1] - ts[0]
    y  = np.full(len(ts), np.nan, dtype = float)
    for n in range(len(ts)):
        if n == 0:
            y[n] = y0
        else:
            y[n] = ((1. - 2. * dt * np.tanh(2. * ts[n])) ** n) * y0
    return y

def BE(ts, y0):
    dt = ts[1] - ts[0]
    y  = np.full(len(ts), np.nan, dtype = float)
    for n in range(len(ts)):
        if n == 0:
            y[n] = y0
        else:
            y[n] = y0 / ((1. + 2. * dt * np.tanh(2. * ts[n])) ** n)
    return y

def TR(ts, y0):
    dt = ts[1] - ts[0]
    y  = np.full(len(ts), np.nan, dtype = float)
    for n in range(len(ts)):
        if n == 0:
            y[n] = y0
        else:
            y[n] = y0 * (((1. - dt * np.tanh(2. * ts[n])) / (1. + dt * np.tanh(2. * ts[n]))) ** n)
    return y

def RK(ts, y0):
    dt = ts[1] - ts[0]
    y  = np.full(len(ts), np.nan, dtype = float)
    for n, t in enumerate(ts):
        if n == 0:
            y[n] = y0
        else:
            q1 = dt * -2. * np.tanh(2 * ts[n]) * y[n - 1]
            phi1 = y[n - 1] + q1
            q2 = -1. * q1
            y[n] = phi1 + q2 / 2.
    return y
```

# ∆t = 0.2

```python
dt = 0.2
ts = np.arange(0., 30. + dt, dt)
y0 = 1.

exact = 1. / np.cosh(2. * ts)

yFE = FE(ts, y0)
yBE = BE(ts, y0)
yTR = TR(ts, y0)
yRK = RK(ts, y0)

fig, sub = plt.subplots(figsize = (10, 5))

sub.plot(ts, exact, c = "k", lw = 2, label = "Exact")
sub.plot(ts, yFE, c = "r"     , lw = 1, label = "FE")
sub.plot(ts, yBE, c = "g"     , lw = 1, label = "BE")
sub.plot(ts, yTR, c = "b"     , lw = 1, label = "TR")
sub.plot(ts, yRK, c = "violet", lw = 1, label = "RK")

sub.set_xlim(0, 30)
sub.set_ylim(-1, 1)

sub.set_xlabel("t", fontsize = 15)
sub.set_ylabel("y", fontsize = 15)

sub.legend()

plt.savefig("./HW3_1/1_1.png", dpi = 500, bbox_inches = "tight")
```

# Plot local error at t=3.0 and t=30.0 with respect to ∆t (log-log plot), and discuss how the numerical error evolves in terms of accuracy and stability.

```python
ts2  = np.arange(0., 30. + 0.2, 0.2)
ts9  = np.arange(0., 30. + 0.9, 0.9)
ts11 = np.arange(0., 30. + 1.1, 1.1)
ts25 = np.arange(0., 30. + 2.5, 2.5)

y0 = 1.

exact2  = 1. / np.cosh(2. * ts2 )
exact9  = 1. / np.cosh(2. * ts9 )
exact11 = 1. / np.cosh(2. * ts11)
exact25 = 1. / np.cosh(2. * ts25)

yFE2 , yBE2 , yTR2 , yRK2  = FE(ts2 , y0), BE(ts2 , y0), TR(ts2 , y0), RK(ts2 , y0)
yFE9 , yBE9 , yTR9 , yRK9  = FE(ts9 , y0), BE(ts9 , y0), TR(ts9 , y0), RK(ts9 , y0)
yFE11, yBE11, yTR11, yRK11 = FE(ts11, y0), BE(ts11, y0), TR(ts11, y0), RK(ts11, y0)
yFE25, yBE25, yTR25, yRK25 = FE(ts25, y0), BE(ts25, y0), TR(ts25, y0), RK(ts25, y0)

ts = [3., 30.]

errorFE2 = np.interp(ts, ts2, yFE2) - np.interp(ts, ts2, exact2)
errorBE2 = np.interp(ts, ts2, yBE2) - np.interp(ts, ts2, exact2)
errorTR2 = np.interp(ts, ts2, yTR2) - np.interp(ts, ts2, exact2)
errorRK2 = np.interp(ts, ts2, yRK2) - np.interp(ts, ts2, exact2)

errorFE9 = np.interp(ts, ts9, yFE9) - np.interp(ts, ts9, exact9)
errorBE9 = np.interp(ts, ts9, yBE9) - np.interp(ts, ts9, exact9)
errorTR9 = np.interp(ts, ts9, yTR9) - np.interp(ts, ts9, exact9)
errorRK9 = np.interp(ts, ts9, yRK9) - np.interp(ts, ts9, exact9)

errorFE11 = np.interp(ts, ts11, yFE11) - np.interp(ts, ts11, exact11)
errorBE11 = np.interp(ts, ts11, yBE11) - np.interp(ts, ts11, exact11)
errorTR11 = np.interp(ts, ts11, yTR11) - np.interp(ts, ts11, exact11)
errorRK11 = np.interp(ts, ts11, yRK11) - np.interp(ts, ts11, exact11)

errorFE25 = np.interp(ts, ts25, yFE25) - np.interp(ts, ts25, exact25)
errorBE25 = np.interp(ts, ts25, yBE25) - np.interp(ts, ts25, exact25)
errorTR25 = np.interp(ts, ts25, yTR25) - np.interp(ts, ts25, exact25)
errorRK25 = np.interp(ts, ts25, yRK25) - np.interp(ts, ts25, exact25)
```

```python
fig, subs = plt.subplots(ncols = 2, figsize = (10, 5))

for tt, t in enumerate(ts):
    sub = subs[tt]
    
    sub.plot([0.2, 0.9, 1.1, 2.5], np.abs([errorFE2[tt], errorFE9[tt], errorFE11[tt], errorFE25[tt]]), c = "r"     , lw = 1 ,label = "FE")
    sub.plot([0.2, 0.9, 1.1, 2.5], np.abs([errorBE2[tt], errorBE9[tt], errorBE11[tt], errorBE25[tt]]), c = "g"     , lw = 1 ,label = "BE")
    sub.plot([0.2, 0.9, 1.1, 2.5], np.abs([errorTR2[tt], errorTR9[tt], errorTR11[tt], errorTR25[tt]]), c = "b"     , lw = 1 ,label = "TR")
    sub.plot([0.2, 0.9, 1.1, 2.5], np.abs([errorRK2[tt], errorRK9[tt], errorRK11[tt], errorRK25[tt]]), c = "violet", lw = 1 ,label = "RK")
    
    sub.legend()
    
    sub.set_xscale("log")
    sub.set_yscale("log")
    
    sub.set_xlabel("Size of h", fontsize = 15)
    sub.set_ylabel("Error", fontsize = 15)
    
plt.savefig("./HW3_1/4.png", dpi = 500, bbox_inches = "tight")
```
